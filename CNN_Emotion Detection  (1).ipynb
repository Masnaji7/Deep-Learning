### Model Building

```python
seq_model = Sequential()
```

### First Convolutional Layer

```python
seq_model.add( Conv2D(128, kernel_size = (3,3), strides = (1,1), 
                      activation = "relu", padding = "same", input_shape = (60,60,3)))
```

```python
seq_model.add( MaxPooling2D(pool_size = (2,2), padding = "same"))
```

```python
seq_model.add( Dropout(0.25))
```

```python
seq_model.add( BatchNormalization(axis = -1))
```

### Second Convolutional Layer

```python
seq_model.add( Conv2D( 256, kernel_size = (3,3), strides = (1,1), 
                      activation = "relu", padding = "same"))
```

```python
seq_model.add( MaxPooling2D(pool_size = (2,2), padding = "same"))
```

```python
seq_model.add( Dropout(0.25))
```

```python
seq_model.add( BatchNormalization(axis = -1))
```

### Third Convolutional Layer

```python
seq_model.add( Conv2D( 128, kernel_size = (3,3), strides = (1,1), 
                      activation = "relu", padding = "same" ))
```

```python
seq_model.add( MaxPooling2D( pool_size = (2,2), padding = "same"))
```

```python
seq_model.add( Dropout(0.25))
```

```python
seq_model.add( BatchNormalization( axis = -1))
```

### Feed Forward Layers

```python
seq_model.add( Flatten())
```

```python
seq_model.add( Dense(512, activation = "relu"))
```

```python
seq_model.add( Dense(1024, activation = "relu"))
```

```python
seq_model.add( Dense( 512, activation = "relu"))
```

```python
seq_model.add( Dense( 7, activation = "softmax"))
```

```python
seq_model.summary()
```

```python
seq_model.compile( optimizer = "adam",
                 loss = "sparse_categorical_crossentropy",
                 metrics = ["accuracy"])
```

```python
seq_model.fit(x_train, y_train, epochs = 20)
```

```python
seq_model.evaluate(x_test,y_test)
```
